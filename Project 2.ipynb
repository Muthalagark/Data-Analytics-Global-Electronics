```python
# executors:
# name: python
# language: python
# setup: "! pip install pandas"
# inputs:
# - name: Stores_1_csv
# - name: Products_1_csv
# - name: Customers_1_csv
# - name: Sales__1__1_csv
# cells:
# - name: load_data
# executor: python
# inputs:
# - name: Stores_1_csv
# - name: Products_1_csv
# - name: Customers_1_csv
# - name: Sales__1__1_csv
#     code: |
import pandas as pd
from io import StringIO
      
# Load the CSV files
stores_df = pd.read_csv('C:/Users/User/Downloads/Stores.csv')
products_df = pd.read_csv('C:/Users/User/Downloads/Products.csv')
customers_df = pd.read_csv('C:/Users/User/Downloads/Customers.csv',encoding='latin1')
sales_df = pd.read_csv('C:/Users/User/Downloads/Sales.csv')

```

```python
a=customers_df.groupby('Birthday')
```

```python
# Check for missing values
missing_values_stores = stores_df.isnull().sum()
missing_values_products = products_df.isnull().sum()
missing_values_customers = customers_df.isnull().sum()
missing_values_sales = sales_df.isnull().sum()
```

```python
missing_values_sales
```

```python
# Convert data types
stores_df['Open Date'] = pd.to_datetime(stores_df['Open Date'], errors='coerce')
sales_df['Order Date'] = pd.to_datetime(sales_df['Order Date'], errors='coerce')
sales_df['Delivery Date'] = pd.to_datetime(sales_df['Delivery Date'], errors='coerce')
customers_df['Birthday'] = pd.to_datetime(customers_df['Birthday'], errors='coerce')
```

```python
if 'Open Date' in stores_df.columns:
    stores_df['Open Date'] = pd.to_datetime(stores_df['Open Date'], errors='coerce')

if 'Order Date' in sales_df.columns:
    sales_df['Order Date'] = pd.to_datetime(sales_df['Order Date'], errors='coerce')

if 'Delivery Date' in sales_df.columns:
    sales_df['Delivery Date'] = pd.to_datetime(sales_df['Delivery Date'], errors='coerce')

if 'Birthday' in customers_df.columns:
    customers_df['Birthday'] = pd.to_datetime(customers_df['Birthday'], errors='coerce')
```

```python
# Outputs for missing values and data types
missing_values = {
    'Stores': missing_values_stores,
    'Products': missing_values_products,
    'Customers': missing_values_customers,
    'Sales': missing_values_sales
}

# Data types
data_types = {
    'Stores': stores_df.dtypes,
    'Products': products_df.dtypes,
    'Customers': customers_df.dtypes,
    'Sales': sales_df.dtypes
}
```

```python
# Merging datasets
merged_df = sales_df.merge(customers_df, on='CustomerKey', how='left') \
                    .merge(stores_df, on='StoreKey', how='left') \
                    .merge(products_df, on='ProductKey', how='left')

# Outputs
merged_df_head = merged_df.head()
```
